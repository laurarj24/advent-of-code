Pair up the numbers and measure how far apart they are. Pair up the smallest number in the left list with the smallest number in the right list, then the second-smallest left number with the second-smallest right number, and so on.

Within each pair, figure out how far apart the two numbers are; you'll need to add up all of those distances. For example, if you pair up a 3 from the left list with a 7 from the right list, the distance apart is 4; if you pair up a 9 with a 3, the distance apart is 6.
To find the total distance between the left list and the right list, add up the distances between all of the pairs you found.

```python
import pandas as pd

df = pd.read_csv("../input.csv", header=None, sep="   ", engine='python') # I create a dataframe in Pandas using the data from the CSV input.
df.columns = ["list 1","list 2"] # I assign the columns names as "list 1" and "list 2"

df
```

```python
df_sort = pd.DataFrame({"list 1 sorted": df['list 1'].sort_values().reset_index(drop=True), "list 2 sorted": df['list 2'].sort_values().reset_index(drop=True)})
df_sort

# I create a new dataframe using the two columns from the first dataframe after sorting them
```

```python
df_sort['Difference'] = abs(df_sort['list 1 sorted'] - df_sort['list 2 sorted'])

# I add a new column that calculates the difference between the sorted values of 'list_1' and 'list_2'.

df_sort
```

```python
df_sort['Difference'].sum() # Using the Pandas sum method, I calculate the total sum of these differences.
```

This time, you'll need to figure out exactly how often each number from the left list appears in the right list. Calculate a total similarity score by adding up each number in the left list after multiplying it by the number of times that number appears in the right list.

```python
count = df_sort['list 1 sorted'].apply(lambda x: (df_sort['list 2 sorted'] == x).sum() * x)
# With the Pandas apply method, I use a lambda function to find the elements in 'list_1' (sorted) that also exist in 'list_2' (sorted), 
# and I multiply each element by the number of times it appears in 'list_2'.
```

```python
count.sum() # Finally, I sum the results of these products
```
